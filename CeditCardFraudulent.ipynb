**Preprocessing The Data**

```python
def cleaning_data(clean):
    clean.drop(["Unnamed: 0",'cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],axis=1, inplace=True)
    clean.dropna()
    return clean
```

```python
cleaning_data(train)
```

```python
cleaning_data(test)
```

```python
train['gender'] = train['gender'].map({'M': 0, 'F': 1})
```

```python
train.info()
```

```python
encoder = LabelEncoder()
def encode(data):
    data["merchant"] = encoder.fit_transform(data["merchant"])
    data["category"] = encoder.fit_transform(data["category"])
    data["gender"] = encoder.fit_transform(data["gender"])
    data["job"] = encoder.fit_transform(data["job"])
    return data
```

```python
encode(train)
```

```python
encode(test)
```

**Data Visualization**

```python
exit_counts = train["is_fraud"].value_counts()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(exit_counts, labels=["NOT FRAUDULENT", "FRADULENT"], autopct="%0.0f%%")
plt.title("COUNT of Frauds")
plt.tight_layout()
plt.show()
```

```python
x = train.drop(columns=["is_fraud"], inplace = False)
y = train["is_fraud"]
```

**Train and Test Split**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**Defining The Model**

```python
model1 = LogisticRegression()
```

```python
columns = ['LogisticRegression']
accuracies=[]
```

```python
def cal (model):
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    accuracy = accuracy_score(y_test, predictions)

    accuracies.append(accuracy)

    print("-----Model Evaluation on Test Data-----")
    print(model)
    print()
    print(accuracy)
```

```python
cal(model1)
```
